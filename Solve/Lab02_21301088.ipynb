```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

Task 1: Zigzag Walk

```python
def walk_zigzag(floor):
    row, column = floor.shape
    for i in range(column):
        if i % 2 == 0:
            for j in range(row):
                if j % 2 == 0:
                    print(floor[j][i], end=" ")
            print()
        else:
            for k in range(row - 1, -1, -1):
                if k % 2 != 0:
                    print(floor[k][i], end=" ")
            print()
    print()


floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

Task 3: Matrix Manipulation

```python
def reverse_Matrix(matrix):
    row, col = matrix.shape
    for i in range(row // 2):
        for j in range(col):
            temp = matrix[i][j]
            matrix[i][j] = matrix[row - 1 - i][col - 1 - j]
            matrix[row - 1 - i][col - 1 - j] = temp
    return matrix

matrix = np.array([
[14,  8,  0,  4],
[9,  8,  13,  13],
[9,  3,  1,  4],
[2,  10,  13,  6]
])
print_matrix(matrix)
print('################')
print()
reversed_matrix = reverse_Matrix(matrix)
print_matrix(reversed_matrix)

#This should print
#|  6  |  13 |  10 |  2  |
#-------------------------
#|  4  |  1  |  3  |  9  |
#-------------------------
#|  13  |  13  |  8 |  9 |
#-------------------------
#|  4 |  0  |  8  |  14  |
#-------------------------
```

Task 5: Matrix Compression

```python
def compress_matrix(matrix):
  row,col=matrix.shape
  arr=np.zeros(((row//2),(col//2)),dtype=object)
  for i in range(0, row, 2):
        for j in range(0, col, 2):
            arr[i // 2, j // 2] = matrix[i,j] + matrix[i,j+1] + matrix[i+1,j] + matrix[i+1,j+1]
  return arr

matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print_matrix(matrix)
print('################')
print()
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print
#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```

Task 6: Game Arena

```python
def play_game(arena):
    row,col = len(arena), len(arena[0])
    point = 0

    for i in range(row):
        for j in range(col):
            if arena[i][j] % 50 == 0 and arena[i][j] > 0:
                for k, l in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                    if 0 <= i + k < row and 0 <= j + l < col:
                        if arena[i + k][j + l] == 2:
                            point += 2

    print(f"Points Gained: {point}.", end=" ")
    print("Your team is out." if point < 10 else "Your team has survived the game.")

arena=np.array([[0,2,2,0],
                [50,1,2,0],
                [2,2,2,0],
                [1,100,2,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 6. Your team is out.

print(".....................")
arena=np.array([[0,2,2,0,2],
                [1,50,2,1,100],
                [2,2,2,0,2],
                [0,200,2,0,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 14. Your team has survived the game.
```
