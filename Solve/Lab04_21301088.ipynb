```python
#Assignment Part 1
class Patient:
    #write a constructor
    def __init__(self, id, name, age, bloodgroup, next=None, prev=None):
        self.id = id
        self.name = name
        self.age = age
        self.bloodgroup = bloodgroup
        self.next = next
        self.prev = prev

```

```python
class WRM:
  #write a constructor
    def __init__(self):
        self.head = None

    def registerPatient(self, id, name, age, bloodgroup):
        new_patient = Patient(id, name, age, bloodgroup)
        if not self.head:
            self.head = new_patient
            new_patient.next = new_patient
            new_patient.prev = new_patient
        else:
            new_patient.prev = self.head.prev
            new_patient.next = self.head
            self.head.prev.next = new_patient
            self.head.prev = new_patient

    def servePatient(self):
        if not self.head:
            print("No patients in the waiting room.")
            return None
        else:
            served_patient = self.head
            if self.head.next == self.head:  # Only one patient in the waiting room
                self.head = None
            else:
                self.head.next.prev = self.head.prev
                self.head.prev.next = self.head.next
                self.head = self.head.next
            print("Serving patient:", served_patient.name)
            return served_patient

    def showAllPatient(self):
        if not self.head:
            print("No patients in the waiting room.")
        else:
            current_patient = self.head
            print("Patients in the waiting room:")
            while True:
                print("ID:", current_patient.id, "- Name:", current_patient.name)
                current_patient = current_patient.next
                if current_patient == self.head:
                    break

    def canDoctorGoHome(self):
        if not self.head:
            print("Yes, the doctor can go home.")
        else:
            print("No, there are patients waiting in the waiting room.")

    def cancelAll(self):
        self.head = None
        print("All appointments canceled. Doctor can go to lunch.")

    def ReverseTheLine(self):
        if not self.head:
            print("No patients in the waiting room.")
        else:
            current_patient = self.head.prev
            while current_patient != self.head:
                current_patient.next, current_patient.prev = current_patient.prev, current_patient.next
                current_patient = current_patient.prev
            self.head.next, self.head.prev = self.head.prev, self.head.next
            self.head = self.head.next
            print("Line reversed. Last patient is now first.")
```

```python
#Write a Tester Code in this cell
print("**Welcome to Waiting Room Management System**")
wr_manager = WRM()
while True:
    print("\nOptions:")
    print("1. Register a patient")
    print("2. Serve a patient")
    print("3. Show all patients")
    print("4. Can doctor go home?")
    print("5. Cancel all appointments")
    print("6. Reverse the line")
    print("7. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        bloodgroup = input("Enter patient blood group: ")
        wr_manager.registerPatient(id, name, age, bloodgroup)
        print("Patient registered successfully.")

    elif choice == '2':
        served_patient = wr_manager.servePatient()
        if served_patient:
            print("Serving patient:", served_patient.name)

    elif choice == '3':
        wr_manager.showAllPatient()

    elif choice == '4':
        wr_manager.canDoctorGoHome()

    elif choice == '5':
        wr_manager.cancelAll()

    elif choice == '6':
        wr_manager.ReverseTheLine()

    elif choice == '7':
        print("Exiting the Waiting Room Management System. Have a nice day!")
        break

    else:
        print("Invalid choice. Please enter a valid option.")
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    e.next = None
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
    diamond_count = 0
    sand_count = 0
    for i in string:
        if i == '<':
            stack.push(i)
        elif i == '>':
            if not stack.isEmpty():
                stack.pop()
                diamond_count += 1
            else:
                sand_count += 1
    return diamond_count




print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):
    if n <= 0 or stack.isEmpty():
        return
    temp = Stack()
    for i in range(n - 1):
        temp.push(stack.pop())
    stack.pop()
    while not temp.isEmpty():
        stack.push(temp.pop())



print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```

Task 3: Stack Reverse

```python
def conditional_reverse(stack):
    if stack.isEmpty():
        return stack
    temp = Stack()
    current = stack.pop()
    temp.push(current)
    while not stack.isEmpty():
        next_element = stack.pop()
        if next_element != temp.peek():
            temp.push(next_element)
        current = next_element
    return temp



print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
